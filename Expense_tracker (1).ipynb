```python
import csv
import datetime
import os
from collections import defaultdict
import matplotlib.pyplot as plt

# Clear all previous records (except the header) every time the program runs
with open("expenses.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Category", "Description", "Amount"])

EXPENSE_FILE = "expenses.csv"

# Create the file if it doesn't exist
if not os.path.exists(EXPENSE_FILE):
    with open(EXPENSE_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Date", "Category", "Description", "Amount"])


def add_expense():
    try:
        date = datetime.date.today().strftime("%Y-%m-%d")
        category = input("Enter category (Food, Transport, Entertainment, etc.): ").strip().title()
        description = input("Enter expense description: ").strip()
        amount = float(input("Enter amount spent: "))

        with open(EXPENSE_FILE, "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([date, category, description, amount])

        print("\n‚úÖ Expense added successfully!\n")
    except ValueError:
        print("\n‚ö†Ô∏è Invalid input! Please enter a valid amount.\n")


def view_expenses():
    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            data = list(reader)

        if len(data) <= 1:
            print("\nüìÇ No expenses recorded yet!\n")
            return

        print("\nüìä Your Expenses:\n")
        for row in data[1:]:
            print(f"{row[0]} | {row[1]} | {row[2]} | ‚Çπ{row[3]}")
        print()
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error reading expenses: {e}\n")


def expense_summary():
    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            data = list(reader)[1:]

        if not data:
            print("\nüìâ No expense data to summarize!\n")
            return

        summary = defaultdict(float)
        for row in data:
            summary[row[1]] += float(row[3])

        print("\nüìå Expense Summary:\n")
        for category, total in summary.items():
            print(f"{category}: ‚Çπ{total:.2f}")
        print()
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error summarizing expenses: {e}\n")


def export_monthly_summary():
    month = input("Enter month to export summary (YYYY-MM): ").strip()
    filename = f"{month}_summary.csv"

    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            data = [row for row in reader if row[0].startswith(month)]

        if len(data) <= 1:
            print("\nüìÇ No expenses found for this month!\n")
            return

        with open(filename, "w", newline="") as out_file:
            writer = csv.writer(out_file)
            writer.writerows(data)

        print(f"\nüì§ Monthly summary exported to {filename}!\n")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error exporting summary: {e}\n")


def filter_expenses_by_month():
    month = input("Enter month to view (YYYY-MM): ").strip()
    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            filtered = [row for row in list(reader)[1:] if row[0].startswith(month)]

        if not filtered:
            print("\nüìÖ No expenses found for that month.\n")
            return

        print(f"\nüìÖ Expenses for {month}:\n")
        for row in filtered:
            print(f"{row[0]} | {row[1]} | {row[2]} | ‚Çπ{row[3]}")
        print()
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error filtering expenses: {e}\n")


def visualize_expenses():
    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            data = list(reader)[1:]

        if not data:
            print("\nüìä No expenses to visualize!\n")
            return

        summary = defaultdict(float)
        for row in data:
            summary[row[1]] += float(row[3])

        categories = list(summary.keys())
        amounts = list(summary.values())

        plt.figure(figsize=(8, 5))
        plt.bar(categories, amounts, color="skyblue")
        plt.xlabel("Category")
        plt.ylabel("Amount Spent (‚Çπ)")
        plt.title("Expense Breakdown by Category")
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error displaying chart: {e}\n")


def check_budget_limits():
    budgets = {}
    while True:
        category = input("Enter category for budget limit (or press enter to stop): ").strip().title()
        if not category:
            break
        try:
            limit = float(input(f"Enter budget limit for {category}: "))
            budgets[category] = limit
        except ValueError:
            print("Invalid amount. Try again.")

    try:
        with open(EXPENSE_FILE, "r") as file:
            reader = csv.reader(file)
            data = list(reader)[1:]

        actuals = defaultdict(float)
        for row in data:
            actuals[row[1]] += float(row[3])

        print("\nüì¢ Budget Alerts:")
        for cat, limit in budgets.items():
            spent = actuals.get(cat, 0)
            print(f"{cat}: Spent ‚Çπ{spent:.2f} / Budget ‚Çπ{limit:.2f}")
            if spent > limit:
                print(f"‚ö†Ô∏è Alert: Budget exceeded in {cat}!\n")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Error checking budgets: {e}\n")


# Main menu
def main():
    while True:
        print("\nüí∞ EXPENSE TRACKER üí∞")
        print("1Ô∏è‚É£  Add Expense")
        print("2Ô∏è‚É£  View All Expenses")
        print("3Ô∏è‚É£  Expense Summary")
        print("4Ô∏è‚É£  Export Monthly Summary")
        print("5Ô∏è‚É£  Filter by Month")
        print("6Ô∏è‚É£  Visualize Expenses")
        print("7Ô∏è‚É£  Check Budget Limits")
        print("8Ô∏è‚É£  Exit")
        choice = input("\nEnter your choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            expense_summary()
        elif choice == "4":
            export_monthly_summary()
        elif choice == "5":
            filter_expenses_by_month()
        elif choice == "6":
            visualize_expenses()
        elif choice == "7":
            check_budget_limits()
        elif choice == "8":
            print("\nüëã Exiting... Have a great day!\n")
            break
        else:
            print("\n‚ö†Ô∏è Invalid choice! Please select a valid option.\n")

if __name__ == "__main__":
    main()
```
